```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# Load the iris dataset from scikit-learn
df_47 = pd.read_csv("/content/loanfile.csv")
df_47.head()
```

```python
df_47.shape
```

```python
df_47.isnull().sum()
```

```python
df = df_47.dropna()
```

```python
df['Loan_Status'].unique()
```

```python
df.shape
```

```python
inputs=df.drop('Loan_Status',axis='columns')
target = df[['Loan_Status']]
```

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_features = LabelEncoder()
```

```python
# transform into numeric data
inputs['Gender_n'] = le_features.fit_transform(inputs['Gender'])
inputs['Married_n'] = le_features.fit_transform(inputs['Married'])
inputs['Education_n'] = le_features.fit_transform(inputs['Education'])
inputs['Self_Employed_n'] = le_features.fit_transform(inputs['Self_Employed'])
inputs['Property_Area_n'] = le_features.fit_transform(inputs['Property_Area'])

#transform float to int
inputs['ApplicantIncome'] = inputs['ApplicantIncome'].round().astype(int)
inputs['CoapplicantIncome'] = inputs['CoapplicantIncome'].round().astype(int)
inputs['LoanAmount'] = inputs['LoanAmount'].round().astype(int)
inputs['Loan_Amount_Term'] = inputs['Loan_Amount_Term'].round().astype(int)
inputs['Credit_History'] = inputs['Credit_History'].round().astype(int)

# Convert 'Dependents' column to numeric, handling '3+'
inputs['Dependents'] = inputs['Dependents'].replace('3+', 3).astype(int)
```

```python
inputs.head()
```

```python
inputs_n = inputs.drop(['Loan_ID','Gender','Married','Education','Self_Employed','Property_Area'],axis='columns')
```

```python
target['Loan_Status_n'] = le_features.fit_transform(target['Loan_Status'])
target_n = target.drop(['Loan_Status'],axis='columns')
```

```python
inputs_n.head()
```

```python
target_n.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    inputs_n, target_n, test_size=0.2, random_state=20)
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_test['Actual']=y_test[['Loan_Status_n']]
```

```python
y_test.drop(['Loan_Status_n'],axis='columns',inplace=True)
y_test['Predicted']=model.predict(X_test)
y_test
```
